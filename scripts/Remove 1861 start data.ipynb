```python
import re

regex = re.compile( "1861\.1\.1\s*?=\s*?\{[\w\W]*?}" )

def remove_1861_data( file ):
    print( "Opening {}".format( file ) )
    text = ""
    
    with open( file ) as f:
        text = f.read()
        
        # Nothing found? Don't do anything
        if len( re.findall( regex, text ) ) == 0:
            print( "    No 1861 scope" )
            return
        
        text = re.sub( regex, "", text )
        print( "    Removed 1861 info" )
    
    with open( file, 'w' ) as f:
        f.write( text )
        print( "    Writing result" )
```

```python
import os
import os.path

def explore_folder( folder ):
    print( "Exploring {}".format( folder ) )
    
    for f in os.listdir( folder ):
        full_name = "{}\\{}".format( folder, f )
        
        if os.path.isfile( full_name ):
            remove_1861_data( full_name )
        else:
            explore_folder( full_name )
```

```python
explore_folder( "..\\CTR\\history" )
```
